```python
import matplotlib.pyplot as plt
import seaborn as sb
fig, axs = plt.subplots(1,1, figsize= (15, 15))
dataplot = sb.heatmap(ds.corr(), cmap="YlGnBu", annot=True)
plt.show()
```

```python
Z = ds.to_numpy()
X, y = Z[:, : -1], Z[:, -1]

from scripts.score import Score
with open(f"dataset_{DS}_best_model") as f_model:
    sc = Score(f_model.read(), X, y)
```

```python
y_hat = sc.expr.eval(X)

```

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
sns.residplot(x= y, y= (y_hat - y))
plt.show()
```

```python
a = sns.regplot(x= y, y= y_hat, color= 'red')
ax1 = a.axes.twinx()
sns.histplot(y, ax= ax1)
sns.histplot(y_hat, ax= ax1, color= 'yellow')
plt.show()
```

```python
for i in range(X.shape[1]):
    sns.regplot(x=X[:, i], y=y)
    plt.show()
```

```python
from xgboost import XGBRegressor as XGB
xgb = XGB(objective="reg:squarederror", )

xgb.fit(X, y)
```

```python
feature_important = xgb.get_booster().get_score(importance_type='weight')
```

```python
import numpy as np
sns.barplot(x = np.arange(X.shape[1]), y=list(feature_important.values()))
```
